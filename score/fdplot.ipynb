```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
speeds=[]
for i in range(5):
    speeds.append(np.load(f'/home/tianyu/code/uoftexp/sumosim/results_small_new/276/{i}Main1305_merge356insspeed2730_1.2_coopsecrmruletest_lcCoop0.6_sim1_freedepart.npy')[0:40])
speeds=np.asarray(speeds)

```

```python
speeds.shape
```

```python
# Since we know the shape of the data, we can create placeholder position and time data.
# Assuming that the detectors are equally spaced and the time steps are equal intervals.

# Placeholder position data for 5 detectors
position_data = np.array([10, 40, 70, 100, 130])  # Positions in meters

# Placeholder time data for 89 time steps
time_data = np.linspace(0, 39, 40)  # Assuming 89 time steps at equal intervals

# Simulating the speed data with the given shape (5, 89) with random values.
# In your case, you should use your actual speed data array.
speed_data = speeds  # Speeds between 10 and 30 m/s

# Create a meshgrid for the positions and times
T, P = np.meshgrid(time_data, position_data)

# Create the heatmap
plt.figure(figsize=(20, 5))
new_labels = [str(x*3) for x in range(0, 41, 5)] # 生成0, 15, 30...的标签
plt.xticks(range(0, 41, 5), new_labels) # 设置新的刻度标签
heatmap = plt.pcolormesh(T, P, speed_data, shading='auto', cmap='viridis', vmin=25, vmax=30)
plt.colorbar(heatmap, label='Speed (m/s)')

plt.xlabel('Time(min)')
plt.ylabel('Detector Position(meter)')
plt.title('Temporal-Spatial Distribution of Speed')
plt.show()
```
